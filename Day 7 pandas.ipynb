# concatenation

```python
import pandas as pd
```

```python
dict1={'Name':['Arham','Ali','Ahmad','Arslan','Usama','Majid','Abbad','Amir','Khurram']
       ,'Marks':[88,56,77,56,44,87,57,95,44]
      ,'Grades':['A','C','B','C','D','A','C','A','D']}
df1=pd.DataFrame(dict1)
```

```python
df1
```

```python
df1['Name_Grades']=df1['Name']+" "+df1['Grades']
df1
```

```python
dict1={'Name':['Arham','Ali','Ahmad','Arslan','Usama','Majid','Abbad','Amir','Khurram']
       ,'T.Marks':[100,100,100,100,100,100,100,100,100,]}
df=pd.DataFrame(dict1)
df
```

```python
 s2=pd.Series([88,56,77,56,44,87,57,95,44])
s2
```

```python
df2=df1.assign(Obtained=s2)
df2
```

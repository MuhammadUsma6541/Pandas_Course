# Merge Pandas DataFrame

```python
data1={'key':['K0','K1','K2','K3','K4','K5'],
       'Name':['Usama','Adan','Ahmad','Ali','Awais','AKmal']}
data2={'key':['K0','K1','K2','K3','K4','K5'],
       'Marks':[11,12,13,14,15,16]}
```

```python
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(left=df1,right=df2,on="key")
```

```python
print(pd.merge(left=df2,right=df1,on="key"))
print("On tells that on which col or row you want to merger for both the same")
```

```python
pd.merge(df1,df2)
```

```python
data1={'key':['K0','K1','K2','K3','K7','K9x'],
       'Name':['Usama','Adan','Ahmad','Ali','Awais','AKmal']}
data2={'key':['K0','K1','K2','K3','K4','K5'],
       'Marks':[11,12,13,14,15,16]}
```

```python
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
pd.merge(df1,df2,on="key",how='outer',indicator="Source Info")
```

```python
data1={'key':['K0','K1','K2','K3','K4','K5'],
       'Name':['Usama','Adan','Ahmad','Ali','Awais','AKmal']}
data2={'key':['K6','K7','K8','K9','K10','K11'],
       'Marks':[11,12,13,14,15,16]}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
data1={'key':['K0','K1','K2','K3','K4','K5'],
       'Name':['Usama','Adan','Ahmad','Ali','Awais','AKmal']}
data2={'key':['K0','K1','K2','K3','K4','K5'],
       'Name':['Usama','Adan','Ahmad','Ali','Awais','AKmal']}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="key",suffixes=('_leftDF','_rightDF'))
```

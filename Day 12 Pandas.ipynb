# Replace Method

```python
data1=pd.read_csv('C:\\Users\\pc\\Downloads\\data_#na.csv')
```

```python
data1
```

```python
data1.replace(to_replace=['Priyang'],value='Usama')
```

```python
data1.replace(np.nan,40)
```

```python
# replace NAN in marks with 40 and FF in Grades
# we have to paas the dic and inside our dict value is also like a dict
data1.replace({"Marks":{np.nan:40},"Grades":{np.nan:"FF"}})
```

```python
# replace data with ffill
data1.replace(np.nan,method="ffill")
```

```python
# at is used ro replace a single value of the col by value pair
data1.at[0,"Grades"]="FF"
data1
```

```python
data1.iat[1,1]
```

```python
data1.iat[1,1]=89
data1
```

```python
data1.at[5,"Marks"]=88
data1
```
